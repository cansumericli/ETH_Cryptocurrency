```python
df.resample('W').mean()
```

I get weekly mean of each columns, this doesn't make sense in that case because there is no reason to do that, I could use sum of the entire time period or the average values of High and Low values.

AGG Method : Accepts map of the columns and aggregation functions that you want to run on that column.

```python
df_agg = df.resample('W').agg({'Close' : 'mean', 'High' : 'max', 'Low' : 'min', 'Volume' : 'sum'})
df_agg
```

```python
sns.distplot(df_agg['Close'])
```

```python
sns.distplot(df_agg['High'])
```

```python
sns.distplot(df_agg['Low'])
```

```python
sns.pairplot(data=df_agg)
```

```python
sns.jointplot(x='High', y='Close', data=df_agg, color = 'orange', kind = 'kde')
```
